```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_curve
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv("Predicting_Depression.csv")
```

```python
df
```

```python
df.drop(["Timestamp","Unnamed: 10","Unnamed: 11"],axis=1,inplace=True)
```

```python
df
```

```python
df.drop(["Profession","Which Social media do you frequently used?"],axis=1,inplace=True)
```

```python
df.rename(columns={"How much time do you spend in Social media?":"spent_Time","What is your main purpose to use social media?":
                   "purpose_Of_use","Do you reveal about your mental breakdown through social media posts?":
                   "Reveal_mental_breakdown","If yes, then what is the reason behind it?":"reason_behind_usingSocialMedia",
                   "Are you connected with any mental health support forum?":"Connected_with_MentalHealth_forum"
                  },inplace=True)
df
```

```python
spentTime = pd.get_dummies(df.spent_Time)
```

```python
concatDf = pd.concat([df,spentTime],axis="columns")
concatDf 
```

```python
le = LabelEncoder()
dfle = concatDf 
dfle.Gender = le.fit_transform(dfle.Gender)
dfle.purpose_Of_use = le.fit_transform(dfle.purpose_Of_use)
dfle.Reveal_mental_breakdown = le.fit_transform(dfle.Reveal_mental_breakdown)
dfle.Connected_with_MentalHealth_forum = le.fit_transform(dfle.Connected_with_MentalHealth_forum)
dfle
```

```python
dfle.drop(["Age","spent_Time"],axis=1,inplace=True)
dfle
```

```python
reason = pd.get_dummies(df.reason_behind_usingSocialMedia)
reason
```

```python
finalDf = pd.concat([dfle,reason],axis="columns")
finalDf
```

```python
finalDf.drop(finalDf.columns[[3,12,13,14]],axis=1,inplace=True)
finalDf
```

```python
finalDf.rename(columns={"I don't have supportive family member.":"No_supportive_familyMember",
                        "I haven't so many real friend, with whom I can share.":"Less_real_Friend"    
},inplace=True)
finalDf
```

```python
X = finalDf[['Gender','purpose_Of_use','Reveal_mental_breakdown','Connected_with_MentalHealth_forum','Almost Never','Frequently',
            'Often','Rarely','No_supportive_familyMember','Less_real_Friend']]
y = finalDf['Result']
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
LRmodel = LogisticRegression()
LRmodel.fit(X_train, y_train)
```

```python
y_predicted = LRmodel.predict(X_test)
y_predicted
```

```python
LRmodel.score(X_test,y_test)

```

```python
confusion_matrix(y_test,y_predicted)
```

```python
precision, recall, thresolds = precision_recall_curve(y_test,y_predicted)
precision
```

```python
recall
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_predicted)
```

```python

thresolds
```
